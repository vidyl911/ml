```
#import the image here and copy its filename
#Change the path name 
path_name="/content/untitled.png"
img2 = cv2.imread(path_name)
img = cv2.imread(path_name, cv2.IMREAD_GRAYSCALE)
_, threshold = cv2.threshold(img, 250, 255, cv2.THRESH_BINARY)
_, contours, _ = cv2.findContours(threshold, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```
#Drawing contours
for cnt in contours:
    approx = cv2.approxPolyDP(cnt, 0.01*cv2.arcLength(cnt, True), True)
    cv2.drawContours(img, [approx], 0, (0), 1)
    x = approx.ravel()[0]
    y = approx.ravel()[1]
```

```
#For cropping, saving and giving boundary limits to currency
idx = 0
for c in contours:
  x, y, w, h = cv2.boundingRect(c)
  if (w > 80 and h > 20) and w is not 0 and y is not 0:
    idx += 1
    new_img = img2[y:y+h, x:x+w]
    cv2_imshow(new_img)
    #if not using colab use cv2.imshow insted of cv2_imshow
    cv2.imwrite(str(idx)+'.png', new_img)
    #download it to your PC
```
