```python
import pandas as pd
import numpy as np
import keras

np.random.seed(2)
```

```python
!wget https://www.dropbox.com/s/li9wo93sho7s7bc/creditcard.csv?dl=0
```

```python
data = pd.read_csv('creditcard.csv')
```

## Data exploration

```python
data.head()
```

## Pre-processing

```python
from sklearn.preprocessing import StandardScaler
data['normalizedAmount'] = StandardScaler().fit_transform(data['Amount'].values.reshape(-1,1))
data = data.drop(['Amount'],axis=1)
```

```python
data.head()
```

```python
data = data.drop(['Time'],axis=1)
data.head()
```

```python
X = data.iloc[:, data.columns != 'Class']
y = data.iloc[:, data.columns == 'Class']
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train = np.array(X_train)
X_test = np.array(X_test)
y_train = np.array(y_train)
y_test = np.array(y_test)
```

Decision tree

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train,y_train.ravel())
decision_tree.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, decision_tree.predict(X_test))

import seaborn as sns
sns.heatmap(cm/np.sum(cm), annot=True, 
            fmt='.2%', cmap='Blues')
```

```python
cm=confusion_matrix(y, decision_tree.predict(np.array(X)))
sns.heatmap(cm/np.sum(cm), annot=True, 
            fmt='.2%', cmap='Blues')
```

### Random Forest

```python

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state=0)
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train,y_train.values.ravel())
random_forest.score(X_test,y_test)
```

```python
cm = confusion_matrix(y_test,random_forest.predict(X_test))
sns.heatmap(cm/np.sum(cm), annot=True, 
            fmt='.2%', cmap='Blues')
```

```python
cm = confusion_matrix(y,random_forest.predict(X))
sns.heatmap(cm/np.sum(cm), annot=True, 
            fmt='.2%', cmap='Blues')
```

## XGBoost

```python
# Fitting XGBoost to the Training set
from xgboost import XGBClassifier
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state=0)
classifier = XGBClassifier()
classifier.fit(X_train, y_train.values.ravel())
cm = confusion_matrix(y_test, classifier.predict(X_test))
sns.heatmap(cm/np.sum(cm), annot=True, 
            fmt='.2%', cmap='Blues')

```

```python
# Applying k-Fold Cross Validation
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train.values.ravel(), cv = 2)
accuracies.mean(),accuracies.std()
```

## Deep neural network

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
model = Sequential([
    Dense(units=16, input_dim = 29,activation='relu'),
    Dense(units=24,activation='relu'),
    Dropout(0.5),
    Dense(20,activation='relu'),
    Dense(24,activation='relu'),
    Dense(1,activation='sigmoid'),
])
model.summary()
```

## Training

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(X_train,y_train,batch_size=15,epochs=5)
```

```python
score = model.evaluate(X_test, y_test)
score
```

```python
cm = confusion_matrix(pd.DataFrame(y_test), model.predict(X_test))
sns.heatmap(cm/np.sum(cm), annot=True, 
            fmt='.2%', cmap='Blues')
```

```python
cm = confusion_matrix(pd.DataFrame(y), model.predict(X))
sns.heatmap(cm/np.sum(cm), annot=True, 
            fmt='.2%', cmap='Blues')
```

## Undersampling

```python
fraud_indices = np.array(data[data.Class == 1].index)
number_records_fraud = len(fraud_indices)
number_records_fraud
```

```python
normal_indices = data[data.Class == 0].index
random_normal_indices = np.random.choice(normal_indices, number_records_fraud, replace=False)
random_normal_indices = np.array(random_normal_indices)
print(len(random_normal_indices))
```

```python
under_sample_indices = np.concatenate([fraud_indices,random_normal_indices])
print(len(under_sample_indices))
```

```python
under_sample_data = data.iloc[under_sample_indices,:]
X_undersample = under_sample_data.iloc[:,under_sample_data.columns != 'Class']
y_undersample = under_sample_data.iloc[:,under_sample_data.columns == 'Class']
X_train, X_test, y_train, y_test = train_test_split(X_undersample,y_undersample, test_size=0.3)
X_train = np.array(X_train)
X_test = np.array(X_test)
y_train = np.array(y_train)
y_test = np.array(y_test)
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(X_train,y_train,batch_size=15,epochs=5)
```

```python
cm = confusion_matrix(pd.DataFrame(y_test), model.predict(X_test))
sns.heatmap(cm/np.sum(cm), annot=True, 
            fmt='.2%', cmap='Blues')
```

```python
cm = confusion_matrix(pd.DataFrame(y), model.predict(X))
sns.heatmap(cm/np.sum(cm), annot=True, 
            fmt='.2%', cmap='Blues')
```
